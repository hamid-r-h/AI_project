```python
from __future__ import unicode_literals
import csv
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from pandas_profiling import ProfileReport
from hazm import *
import re 
from sklearn.feature_extraction.text import CountVectorizer
from parsivar import FindStems
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
train, test = pd.read_csv('dataset/train.csv'), pd.read_csv('dataset/inference.csv')
print("train shape: {}, test shape: {}".format(train.shape, test.shape))
train = train.drop_duplicates(subset=['query'])
test.head(10)
```

```python
train_data = train.to_numpy()
test_data = test.to_numpy()
X, y = train_data[:, 1] , train_data[:, 2]
X_test= test_data[:, 1]
```

```python
def preprocess_data(X):
    stop_words = stopwords_list()
    for ide in range(X.shape[0]):
        text = X[ide]
        normalizer = Normalizer(persian_numbers=False)
        text = normalizer.normalize(text)
        my_stemmer = FindStems()
        text=my_stemmer.convert_to_stem(text)
        text = re.sub('[0-9]',' ', text)
        text = re.sub('[۰-۹]',' ', text)
        text = re.sub('[A-Za-z]',' ', text)
        text=re.sub(r'[^\w\s]', ' ',text)
        text= ' '.join(text.split())
        text_words = word_tokenize(text)
        f_text = []
        for current_word in text_words:
            if current_word not in stop_words:
                #if(len(current_word)>3):
                    f_text.append(current_word)
        X[ide] = ' '.join(f_text)
    return X    
```

```python
def counter_x(X,y):
    dff=pd.DataFrame()
    dff['query']=X
    dff['label']=y
    list_rep=[]
    list_x=[]
    list_p=[]
    a = [row['query'] for index,row in dff.iterrows() if row['label'] ==1]
    list_p.append(len(a)/len(dff))
    one = CountVectorizer()
    list_x.append(one)
    list_rep.append(one.fit_transform(a).toarray())
    
    b = [row['query'] for index,row in dff.iterrows() if row['label'] ==2]
    list_p.append(len(b)/len(dff))
    two = CountVectorizer()
    list_x.append(two)
    list_rep.append(two.fit_transform(b).toarray())
    
    c = [row['query'] for index,row in dff.iterrows() if row['label'] ==3]
    list_p.append(len(c)/len(dff))
    three = CountVectorizer()
    list_x.append(three)
    list_rep.append(three.fit_transform(c).toarray())

    d = [row['query'] for index,row in dff.iterrows() if row['label'] ==4]
    list_p.append(len(d)/len(dff))
    four = CountVectorizer()
    list_x.append(four)
    list_rep.append(four.fit_transform(d).toarray())


    e = [row['query'] for index,row in dff.iterrows() if row['label'] ==5]
    list_p.append(len(e)/len(dff))
    five = CountVectorizer()
    list_x.append(five)
    list_rep.append(five.fit_transform(e).toarray())
    return list_x,list_p,list_rep
```

```python
def sum_x(list_x,list_p,list_rep):
    list_count=[]
    list_feature=[]
    list_feature.append(list_x[0].get_feature_names())    
    list_count.append(list_rep[0].sum(axis=0)) 
    one=dict(zip(list_feature[0],list_count[0]))
    
    list_feature.append(list_x[1].get_feature_names())    
    list_count.append(list_rep[1].sum(axis=0))
    two=dict(zip(list_feature[1],list_count[1]))
    
    list_feature.append(list_x[2].get_feature_names())    
    list_count.append(list_rep[2].sum(axis=0))
    three=dict(zip(list_feature[2],list_count[2]))
    
    list_feature.append(list_x[3].get_feature_names())    
    list_count.append(list_rep[3].sum(axis=0))
    four=dict(zip(list_feature[3],list_count[3]))

    list_feature.append(list_x[4].get_feature_names())    
    list_count.append(list_rep[4].sum(axis=0))
    five=dict(zip(list_feature[4],list_count[4]))
    
    return list_count,list_feature,one,two,three,four,five
```

```python
def p_word(list_count,list_feature):
    p_one=[]
    for w,c in zip(list_feature[0],list_count[0]):
        siz=len(list_feature[0])
        p_one.append(c/siz)
    p_two=[]
    for w,c in zip(list_feature[1],list_count[1]):
        siz=len(list_feature[1])
        p_two.append(c/siz)
    p_three=[]
    for w,c in zip(list_feature[2],list_count[2]):
        siz=len(list_feature[2])
        p_three.append(c/siz)
    p_four=[]
    for w,c in zip(list_feature[3],list_count[3]):
        siz=len(list_feature[3])
        p_four.append(c/siz)
    p_five=[]
    for w,c in zip(list_feature[4],list_count[4]):
        siz=len(list_feature[4])
        p_five.append(c/siz)
    dict(zip(list_feature[2],p_three))
    return p_one,p_two,p_three,p_four,p_five
```

```python
def total(list_count,X):
    dff=pd.DataFrame()
    dff['query']=X
    q= [row['query'] for index,row in dff.iterrows()]
    all_v = CountVectorizer()
    z = all_v.fit_transform(q)
    totall = len(all_v.get_feature_names())
    count_all=[]
    count_all.append(list_count[0].sum(axis=0))
    count_all.append(list_count[1].sum(axis=0))
    count_all.append(list_count[2].sum(axis=0))
    count_all.append(list_count[3].sum(axis=0))
    count_all.append(list_count[4].sum(axis=0))
    return count_all,totall
```

```python
def naive_b(test,totall,count_all,list_feature,list_count,list_p,one,two,three,four,five):
    i=-1;
    for q in test['query']:
        i=i+1
        t_word= word_tokenize(q)
        p_one_ls=[]
        p_two_ls=[]
        p_three_ls=[]
        p_four_ls=[]
        p_five_ls=[]
        for w in t_word:
            if w in one.keys():
                count=one[w]
                p_one_ls.append((count + 1)/(count_all[0] + totall))
            else:
                count = 0
                p_one_ls.append((count + 1)/(count_all[0] + totall))
            if w in two.keys():
                count =two[w]
                p_two_ls.append((count + 1)/(count_all[1] + totall))
            else:
                count = 0
                p_two_ls.append((count + 1)/(count_all[1] + totall))
            if w in three.keys():
                count =three[w]
                p_three_ls.append((count + 1)/(count_all[2] + totall))
            else:
                count = 0
                p_three_ls.append((count + 1)/(count_all[2] + totall))
            if w in four.keys():
                count = four[w]
                p_four_ls.append((count + 1)/(count_all[3] + totall))
            else:
                count = 0
                p_four_ls.append((count + 1)/(count_all[3] + totall))
            if w in five.keys():
                count =five[w]
                p_five_ls.append((count + 1)/(count_all[4] + totall))
            else:
                count = 0
                p_five_ls.append((count + 1)/(count_all[4] + totall))
        pro_one=list_p[0]*np.prod(p_one_ls)
        pro_two=list_p[1]*np.prod(p_two_ls)
        pro_three=list_p[2]*np.prod(p_three_ls)
        pro_four=list_p[3]*np.prod(p_four_ls)
        pro_five=list_p[4]*np.prod(p_five_ls)
        res=max(pro_one,pro_two,pro_three,pro_four,pro_five)
        if(res==pro_one):
            test.at[i,'label']=1
        elif(res==pro_two):
            test.at[i,'label']=2
        elif(res==pro_three):
            test.at[i,'label']=3
        elif(res==pro_four):
            test.at[i,'label']=4
        elif(res==pro_five):
            test.at[i,'label']=5            
    return test
```
